```python
# data_preprocessing_diagnostics.py

import numpy as np
import pandas as pd
import seaborn as sns; sns.set(style='whitegrid')
import matplotlib.pyplot as plt
from sklearn.preprocessing import (
    StandardScaler, MinMaxScaler, RobustScaler, Normalizer, Binarizer,
    LabelEncoder, OneHotEncoder, PolynomialFeatures, FunctionTransformer
)
from sklearn.impute import SimpleImputer
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, HashingVectorizer

# ----------------------------------------------------
# Helper functions
# ----------------------------------------------------
def plot_distribution(df, title):
    df.hist(bins=30, figsize=(10, 4))
    plt.suptitle(title, fontsize=14)
    plt.tight_layout()
    plt.show()

def plot_boxplot(df, title):
    plt.figure(figsize=(8, 4))
    sns.boxplot(data=df)
    plt.title(title, fontsize=14)
    plt.show()

def describe_and_plot(df, label):
    print(f"{label} - Description:\n", df.describe(), "\n")
    plot_distribution(df, f"{label} - Distribution")
    plot_boxplot(df, f"{label} - Boxplot")

# ----------------------------------------------------
# 1. StandardScaler - Best for Normal Distribution
# ----------------------------------------------------
df = pd.DataFrame({
    'x1': np.random.normal(0, 2, 1000),
    'x2': np.random.normal(5, 3, 1000),
    'x3': np.random.normal(-5, 5, 1000)
})
describe_and_plot(df, "Original Data for StandardScaler")
ss = StandardScaler()
df_scaled = pd.DataFrame(ss.fit_transform(df), columns=df.columns)
describe_and_plot(df_scaled, "After StandardScaler")

# ----------------------------------------------------
# 2. MinMaxScaler - Use when data needs to be between 0 and 1
# Especially for Neural Networks
# ----------------------------------------------------
df = pd.DataFrame({
    'x1': np.random.chisquare(8, 1000),          # Skewed
    'x2': np.random.beta(8, 2, 1000) * 40,        # Skewed [0, 40]
    'x3': np.random.normal(50, 3, 1000)           # Gaussian
})
describe_and_plot(df, "Original Data for MinMaxScaler")
mm = MinMaxScaler()
df_scaled = pd.DataFrame(mm.fit_transform(df), columns=df.columns)
describe_and_plot(df_scaled, "After MinMaxScaler")

# ----------------------------------------------------
# 3. RobustScaler - Good when Outliers are Present
# ----------------------------------------------------
df = pd.DataFrame({
    'x1': np.concatenate([np.random.normal(20, 1, 1000), np.random.normal(1, 1, 20)]),
    'x2': np.concatenate([np.random.normal(30, 1, 1000), np.random.normal(60, 1, 20)]),
})
describe_and_plot(df, "Original Data with Outliers (RobustScaler)")
robust = RobustScaler()
df_robust = pd.DataFrame(robust.fit_transform(df), columns=df.columns)
describe_and_plot(df_robust, "After RobustScaler")

# ----------------------------------------------------
# 4. Normalizer - Use when you want unit norm vectors
# Best for Text Data or Clustering Cosine Similarities
# ----------------------------------------------------
df = pd.DataFrame({
    'x': np.random.randint(-100, 100, 500).astype(float),
    'y': np.random.randint(-50, 50, 500).astype(float),
    'z': np.random.randint(-150, 150, 500).astype(float),
})
describe_and_plot(df, "Original Data for Normalizer")
normalizer = Normalizer()
df_norm = pd.DataFrame(normalizer.fit_transform(df), columns=df.columns)
describe_and_plot(df_norm, "After Normalizer")

# ----------------------------------------------------
# 5. Binarizer - Use for thresholding (e.g., image processing, flags)
# ----------------------------------------------------
X = np.array([[1., -1., 2.],
              [2., 0., 0.],
              [0., 1., -1.]])
print("Original for Binarizer:\n", X)
binarizer = Binarizer(threshold=0.0)
X_bin = binarizer.fit_transform(X)
print("After Binarization (threshold=0):\n", X_bin)

# ----------------------------------------------------
# 6. Encoding - Ordinal and Nominal
# ----------------------------------------------------
df = pd.DataFrame({
    'Age': [33, 44, 22, 44, 55, 22],
    'Income': ['Low', 'Low', 'High', 'Medium', 'Medium', 'High']
})
df['Income_Encoded'] = df['Income'].map({'Low': 1, 'Medium': 2, 'High': 3})
sns.barplot(x='Income', y='Income_Encoded', data=df)
plt.title("Ordinal Encoding: Income")
plt.show()

# Nominal Encoding: Gender
df = pd.DataFrame({
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Male']
})
le = LabelEncoder()
df['gender_tf'] = le.fit_transform(df['Gender'])
ohe = OneHotEncoder()
one_hot = ohe.fit_transform(df[['gender_tf']]).toarray()
print("One-Hot Encoding:\n", one_hot)

# ----------------------------------------------------
# 7. Imputation - Filling Missing Values
# ----------------------------------------------------
df = pd.DataFrame({
    'A': [1, 2, 3, 4, np.nan, 7],
    'B': [3, 4, 1, np.nan, 4, 5]
})
print("Original with Missing:\n", df)
imputer = SimpleImputer(strategy='median')
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=['A', 'B'])
print("After Median Imputation:\n", df_imputed)

# ----------------------------------------------------
# 8. Polynomial Features - Add interactions and powers
# ----------------------------------------------------
df = pd.DataFrame({'A': [1, 2, 3], 'B': [2, 3, 4]})
pol = PolynomialFeatures(degree=2)
df_poly = pd.DataFrame(pol.fit_transform(df), columns=pol.get_feature_names_out())
print("Polynomial Features:\n", df_poly)

# ----------------------------------------------------
# 9. Custom Transformer Example
# ----------------------------------------------------
def mapping(x):
    x['Age'] = x['Age'] + 2
    x['Score'] = x['Score'] * 10
    return x

df = pd.DataFrame({
    'Age': [33, 44, 22],
    'Score': [3.5, 4.0, 2.0]
})
ft = FunctionTransformer(mapping, validate=False)
df_transformed = ft.fit_transform(df)
print(" Custom Transformation:\n", df_transformed)

# ----------------------------------------------------
# 10. Text Processing Vectorizers
# ----------------------------------------------------
corpus = [
    'This is the first document.',
    'This document is the second document.',
    'And the third one.',
    'Is this the first document?',
]
cv = CountVectorizer()
cv_matrix = cv.fit_transform(corpus).toarray()
print("CountVectorizer:\n", pd.DataFrame(cv_matrix, columns=cv.get_feature_names_out()))

tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(corpus).toarray()
print("TF-IDF:\n", pd.DataFrame(tfidf_matrix, columns=tfidf.get_feature_names_out()))

hv = HashingVectorizer(n_features=5)
hv_matrix = hv.fit_transform(corpus).toarray()
print("HashingVectorizer:\n", hv_matrix)

```
