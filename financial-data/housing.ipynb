## Finance Dataset 
Demostrate Data Analytics to a Financial Issue in Housing

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load dataset
data = fetch_california_housing()

# Create a DataFrame from the dataset
df = pd.DataFrame(data.data, columns=data.feature_names)

# Print the first few rows of the DataFrame
print("First few rows of the California housing dataset:")
print(df.head())
```

```python
# Preview and Describe the dataset
print("\nDataset Information:")
print(df.info())

print("\nDataset Description:")
print(df.describe())
```

```python
# Duplicates
df = df.drop_duplicates()
print("\nAfter removing duplicates:")
print(df.head())    
```

```python
# Standardize the data using StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X, y = data.data, data.target
df_scaled = scaler.fit_transform(X)

print("\nAfter standardizing the data:")
print(df_scaled[:5])  # Print first 5 rows of scaled data
```

```python
# Handle missing values
# Check for missing values
print("\nMissing values in the dataset:")   
print(df.isnull().sum())
```

```python
# Visualize the data
sns.scatterplot(data=df, x='MedInc', y='HouseAge', hue='MedHouseVal', palette='viridis')
plt.title('Scatter Plot of Median Income vs House Age')
plt.xlabel('Median Income')
plt.ylabel('House Age')
```
